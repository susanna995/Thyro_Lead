```python
import pandas as pd
csv_path = 'newboth.csv'
df = pd.read_csv(csv_path)
df
```

```python
df.fillna('?', inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
# Create a LabelEncoder object
label_encoder = LabelEncoder()

df['target'] = label_encoder.fit_transform(df['target'])
# df['sex'] = label_encoder.fit_transform(df['sex'])
df['on_thyroxine'] = label_encoder.fit_transform(df['on_thyroxine'])
df['query_on_thyroxine'] = label_encoder.fit_transform(df['query_on_thyroxine'])
df['on_antithyroid_meds'] = label_encoder.fit_transform(df['on_antithyroid_meds'])
df['sick'] = label_encoder.fit_transform(df['sick'])
df['pregnant'] = label_encoder.fit_transform(df['pregnant'])
df['thyroid_surgery'] = label_encoder.fit_transform(df['thyroid_surgery'])
df['I131_treatment'] = label_encoder.fit_transform(df['I131_treatment'])
df['query_hypothyroid'] = label_encoder.fit_transform(df['query_hypothyroid'])


df['lithium'] = label_encoder.fit_transform(df['lithium'])
df['goitre'] = label_encoder.fit_transform(df['goitre'])
df['tumor'] = label_encoder.fit_transform(df['tumor'])
df['hypopituitary'] = label_encoder.fit_transform(df['hypopituitary'])
df['psych'] = label_encoder.fit_transform(df['psych'])
df['query_hyperthyroid'] = label_encoder.fit_transform(df['query_hyperthyroid'])

df
```

```python
df.replace('F', '1', inplace=True)
df.replace('M', '2', inplace=True)
# df.replace('?', '2', inplace=True)
df['sex'] = df['sex'].str.replace('?', '2')
df
```

```python


# Columns to impute
columns_to_impute = ['T3', 'TSH', 'TT4', 'T4U', 'FTI', 'age']

# Replace '?' with NaN and convert columns to numeric
df[columns_to_impute] = df[columns_to_impute].replace('?', '0') #.apply(pd.to_numeric, errors='coerce')
```

```python
y = df['target']

x = df.drop('target',axis=1)

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=100)
```

```python
import joblib #trained model storage
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Create a Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42) #randomseed(to generte the same test set)  100 descision trees

# Train the model on the training set
rf_classifier.fit(x_train, y_train)

# Make predictions on the testing set
y_pred = rf_classifier.predict(x_test)

# Save the trained model to a .pkl file using joblib
joblib.dump(rf_classifier, 'hyper_hypo_model.pkl')

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

# Print the results
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(classification_rep)
```
